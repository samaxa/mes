```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Szanrio 1
df_1p_germany = pd.read_csv(f'./Results_scenario_1/Kosten_1p_germany.csv', index_col=0)["1p"]
df_2p_germany = pd.read_csv(f'./Results_scenario_1/Kosten_2p_germany.csv', index_col=0)["2p"]
df_3p_germany = pd.read_csv(f'./Results_scenario_1/Kosten_3p_germany.csv', index_col=0)["3p"]

df_1p_sweden = pd.read_csv(f'./Results_scenario_1/Kosten_1p_sweden.csv', index_col=0)["1p"]
df_2p_sweden = pd.read_csv(f'./Results_scenario_1/Kosten_2p_sweden.csv', index_col=0)["2p"]
df_3p_sweden = pd.read_csv(f'./Results_scenario_1/Kosten_3p_sweden.csv', index_col=0)["3p"]

df_1p_egypt = pd.read_csv(f'./Results_scenario_1/Kosten_1p_egypt.csv', index_col=0)["1p"]
df_2p_egypt = pd.read_csv(f'./Results_scenario_1/Kosten_2p_egypt.csv', index_col=0)["2p"]
df_3p_egypt = pd.read_csv(f'./Results_scenario_1/Kosten_3p_egypt.csv', index_col=0)["3p"]
```

```python

kosten_df_1 = pd.DataFrame({
    "Germany 1p": df_1p_germany,
    "Germany 2p": df_2p_germany,
    "Germany 3p": df_3p_germany,
    "Sweden 1p": df_1p_sweden,
    "Sweden 2p": df_2p_sweden,
    "Sweden 3p": df_3p_sweden,
    "Egypt 1p": df_1p_egypt,
    "Egypt 2p": df_2p_egypt,
    "Egypt 3p": df_3p_egypt
})

labels = kosten_df_1.index.tolist()
farben = ["gold", "green", "yellow", "blue", "cyan", "red", "magenta"]

# X-Positionen mit Lücken zwischen Standorten
n_scenarios_per_location = 3
gap = 0.4  # Abstand zwischen Standorten
x = []
for i in range(len(kosten_df_1.columns)):
    loc_idx = i // n_scenarios_per_location  # Standortgruppe
    pos = i + loc_idx * gap
    x.append(pos)
x = np.array(x)

# Gestapelte Säulen plotten
fig, ax = plt.subplots(figsize=(12,6))
bottom = np.zeros(len(kosten_df_1.columns))

for i, component in enumerate(kosten_df_1.index):
    ax.bar(
        x,
        kosten_df_1.loc[component],
        bottom=bottom,
        color=farben[i % len(farben)],
        label=component
    )
    bottom += kosten_df_1.loc[component]

# Achsen, Titel, Legende
ax.set_xticks(x)
ax.set_xticklabels(kosten_df_1.columns, rotation=45, ha="right", fontsize=13)
ax.set_ylabel("Kosten [€]", fontsize=15)

ax.set_title("Szenario 1: Gesamtkosten/Jahr nach Standort bei einem Demand von 1-3 Personen", fontsize=17)
ax.legend(loc="upper right", fontsize=15)

# Gesamtkosten pro Szenario (Summe der Komponenten)
totals = kosten_df_1.sum(axis=0)

# Werte über den Balken schreiben
for i, total in enumerate(totals):
    ax.text(
        x[i],                 # X-Position (Mitte des Balkens)
        total + total*0.01,   # Y-Position leicht oberhalb
        f"{round(total,0):,.0f}€",  # Gerundeter Wert mit Tausenderpunkten
        ha="center", va="bottom", fontsize=11
    )

plt.tight_layout()
plt.savefig("./Results_scenario_1/Szenario_1_Diagramm.png", dpi=300)
plt.show()
```

```python
# Alle Daten in einem DataFrame zusammenfassen
kosten_df_2 = pd.DataFrame({
    "Germany 120": df_120_germany,
    "Germany 140": df_140_germany,
    "Germany 160": df_160_germany,
    "Germany 180": df_180_germany,
    "Sweden 120": df_120_sweden,
    "Sweden 140": df_140_sweden,
    "Sweden 160": df_160_sweden,
    "Sweden 180": df_180_sweden,
    "Egypt 120": df_120_egypt,
    "Egypt 140": df_140_egypt,
    "Egypt 160": df_160_egypt,
    "Egypt 180": df_180_egypt
})

labels = kosten_df_2.index.tolist()
farben = ["gold", "green", "yellow", "blue", "cyan", "red", "magenta"]

# X-Positionen mit Lücken zwischen Standorten
n_scenarios_per_location = 4  # jetzt 120, 140, 160, 180 pro Land
gap = 0.4                    # Abstand zwischen Ländern
x = []
for i in range(len(kosten_df_2.columns)):
    loc_idx = i // n_scenarios_per_location  # Standortgruppe
    pos = i + loc_idx * gap
    x.append(pos)
x = np.array(x)

# Gestapelte Säulen plotten
fig, ax = plt.subplots(figsize=(12,6))
bottom = np.zeros(len(kosten_df_2.columns))

for i, component in enumerate(kosten_df_2.index):
    ax.bar(
        x,
        kosten_df_2.loc[component],
        bottom=bottom,
        color=farben[i % len(farben)],
        label=component
    )
    bottom += kosten_df_2.loc[component]

# Gesamtkosten pro Szenario (Summe der Komponenten)
totals = kosten_df_2.sum(axis=0)

# Achsen, Titel, Legende
ax.set_xticks(x)
ax.set_xticklabels(kosten_df_2.columns, rotation=45, ha="right", fontsize=14)
ax.set_ylabel("Kosten [€]", fontsize=15)
ax.set_ylim(0, max(totals) * 1.12)  # etwas Luft nach oben
ax.set_title("Szenario 2: Gesamtkosten/Jahr nach Standort bei einer Anbaufläche von 120–180 m2", fontsize=18)
ax.legend(loc="upper right", fontsize=15)

# Werte über den Balken schreiben
for i, total in enumerate(totals):
    ax.text(
        x[i],
        total + total*0.01,
        f"{round(total,0):,.0f}€",
        ha="center", va="bottom", fontsize=11
    )

plt.tight_layout()
plt.savefig("./Results_scenario_2/Szenario_2_Diagramm.png", dpi=300)
plt.show()
```
